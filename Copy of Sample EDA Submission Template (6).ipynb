````python
#6.in which quarter of time the booking got higher than usual in which part of location.
# define a function to know quarter
def quarter(string):
    if string[5:7] in ('01','02','03'):
      return('Q1')
    elif string[5:7] in ('04','05','06'):
      return('Q2')
    elif string[5:7] in ('07','08','09'):
      return('Q3')
    elif string[5:7] in ('10','11','12'):
      return('Q4')
    else:
      return('no quater')
 #not null value in last review column     
no_na=df.loc[~df['last_review'].isna()]

#create a new column named quarter
no_na['quarter']=no_na['last_review'].apply(quarter)


no_na.groupby(['neighbourhood','quarter'])['number_of_reviews'].sum()
````

````python
7.#In which Neighbourhood how many hosts are present. 
hosts_in_each_neighbourhood=df.groupby('neighbourhood')['host_name'].count()
hosts_in_each_neighbourhood.sort_values(ascending=False)
````

````python
# host count in each Area.
hosts_in_each_neighbourhood=df.groupby('neighbourhood_group')['host_name'].count()
hosts_in_each_neighbourhood.sort_values(ascending=False)
````

````python
#host count of a perticular room type in a perticular Area.
hosts_in_each_neighbourhood=df.groupby(['neighbourhood_group','room_type'])['host_name'].count()
hosts_in_each_neighbourhood.sort_values(ascending=False)
````

````python
#Aerage price in each Area
hosts_in_each_neighbourhood=df.groupby('neighbourhood_group')['price'].mean()
hosts_in_each_neighbourhood.sort_values(ascending=False)
````

````python
#8.Room_types and their relation with availability in different neighbourhood groups?
df.groupby(['room_type','neighbourhood_group'])['availability_365'].value_counts()
````

````python
# count of availabilty of 365 days in each Area
days_365 = df[df['availability_365']==365]
days_365.groupby('neighbourhood_group')['availability_365'].value_counts()
````
